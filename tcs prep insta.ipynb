```python
n = int(input())
arr = list(map(int, input().split()))

m = float('-inf')   # start with very small number
c = 0               # counter

for a in arr:       # loop through each number
    if a > m:       # if it's a new maximum
        m = a
        c += 1

print(c)
```

```python
import sys

n = int(input())        # number of inputs
c = 0                   # counter for new maximums
m = -sys.maxsize - 1    # initialize with very small number

for _ in range(n):      # loop exactly n times
    a = int(input())    # read next number
    if a > m:           # check if it's a new maximum
        m = a           # update maximum
        c += 1          # increment counter

print(c)
```

```python
str = input()
n = int(input())
max_val = 0
count = 0
for i in range(len(str)):
    if i % n == 0:
        max_val = max(count, max_val)
        count = 0
    if str[i] == 'a':
        count += 1
if count > max_val:
    max_val = count
print(max_val)
```

```python
n = int(input())
fact = [0] * (n + 1)
fact[0] = 1
for i in range(1, n + 1):
    fact[i] = fact[i - 1] * i
print(fact[n - 1] * 2)
```

```python
n = input("Enter a number: ")
r = int(input("Enter multiplier: "))

# Step 1: sum of digits of n
t = sum(int(i) for i in n)

# Step 2: multiply by r
p = t * r

# Step 3: sum of digits of the product
result = sum(int(i) for i in str(p))

print(result)
```

```python
n = int(input())
arr = list(map(int, input().split()))
d, x = map(int, input().split())
countEven = 0
countOdd = 0
for i in range(n):
    if arr[i] % 2 == 0:
        countEven += 1
    else:
        countOdd += 1
if d % 2 != 0:
    if countEven == 0:
        print("0")
    else:
        print(countEven * x)
else:
    if countOdd == 0:
        print("0")
    else:
        print(countOdd * x)
```
